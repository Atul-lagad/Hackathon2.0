```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("imdb_top_1000.csv")
data.head()
```

```python
# Data:-
# Poster_Link - Link of the poster that imdb using
# Series_Title = Name of the movie
# Released_Year - Year at which that movie released
# Certificate - Certificate earned by that movie
# Runtime - Total runtime of the movie
# Genre - Genre of the movie
# IMDB_Rating - Rating of the movie at IMDB site
# Overview - mini story/ summary
# Meta_score - Score earned by the movie
# Director - Name of the Director
# Star1,Star2,Star3,Star4 - Name of the Stars
# No_of_votes - Total number of votes
# Gross - Money earned by that movie
```

## **Data Cleansing**

```python
data.info()
```

```python
#mean of metascore to fill nill values
meta_score_mean=data['Meta_score'].mean()

#the datatype of Gross column is object ,converting ii into integer and getting mean of it
data['Gross']=data['Gross'].str.replace(',', '').astype(float)
gross_mean=data['Gross'].mean()
```

```python
# Filling Null values of meta score and gross

data['Meta_score'].fillna(meta_score_mean,inplace=True)
data['Gross'].fillna(gross_mean,inplace=True)
```

```python
#to fill values of Certificate are not integer , so for string Forward Fill is good

data['Certificate'] = data['Certificate'].fillna(method='ffill')
```

```python
data.info()
```

## **Data Exploration**

**Director's Impact on Gross Earnings**

```python
top_avg = data.groupby(['Genre', 'Director'])['Gross'].mean().reset_index()
top_avg
```

Movies in Genre of Action and Adventure have highest gross, and Director "Christopher Nolan" haev highest avg gross in movies

**Year wise relesed movie count**

```python

top_movie_relesed_year=data['Released_Year'].value_counts().head(10)
top_movie_relesed_year
```

```python
sns.relplot(x=top_movie_relesed_year.index,y=top_movie_relesed_year.values,kind='line')
```

**Genre wise avg IMDB rating**




```python

average_ratings = data.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).reset_index()
average_ratings=average_ratings.head(10)
average_ratings
```

```python
bar=sns.barplot(x=average_ratings['IMDB_Rating'],y=average_ratings['Genre'])

for i in bar.containers:
    bar.bar_label(i,)

plt.title('Average IMDB Ratings by Genre')
plt.xlabel('Average IMDB Rating')
plt.ylabel('Genre')
```

**Genre wise avg Gross**

```python
average_gross = data.groupby('Genre')['Gross'].mean().sort_values(ascending=False).reset_index()
average_gross=average_gross.head(10)
average_gross
```

```python
bar=sns.barplot(x=average_gross['Gross'],y=average_ratings['Genre'])

for i in bar.containers:
    bar.bar_label(i,)

plt.title('Average Gross by Genre')
plt.xlabel('Average gross')
plt.ylabel('Genre')
```

**Impact of Movie Length on Ratings:**

```python
sns.scatterplot(x='Runtime', y='IMDB_Rating', data=data)
plt.title('Impact of Movie Length on IMDb Ratings')
plt.xlabel('Movie Length (minutes)')
plt.ylabel('IMDb Rating')
plt.show()
```

It conclude that avg movie time 150 have highest IMDB rating




**Impact of Movie Length on Earnings:**


```python
sns.scatterplot(x='Runtime', y='Gross', data=data)
plt.title('Impact of Movie Length on Gross Earnings')
plt.xlabel('Movie Length (minutes)')
plt.ylabel('Gross Earnings')
plt.show()
```

In gross earning also movies length with 150 have highest earning , while runtime 100 and 200 have avg gross earning
